```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
complete_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

complete_data.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#total schools
total_schools= len(complete_data["school_name"].unique())

#Total students
total_students = complete_data["Student ID"].count()

#Total Budget 
total_budget=school_data["budget"].sum()

#Average Math score
avg_math_score =complete_data["math_score"].mean()

#Average Reading Score 
avg_read_score =complete_data["reading_score"].mean()

#overall passing scores
overall_pass= (avg_math_score+ avg_read_score)/2
avg_math_score

math_pass=complete_data[(complete_data["math_score"] >= 70)].count()["student_name"]
#percent_pass_math= (math_pass/total_students)*100
percent_pass_math=math_pass / float(total_students) * 100


# %passing reading 
read_pass=complete_data[(complete_data["reading_score"] >= 70)].count()["student_name"]
percent_pass_read= (read_pass/total_students)*100

#overall passing percent 
overall_percent_pass= (percent_pass_math+percent_pass_read)/2

District_Summary = pd.DataFrame({"Total Schools": [total_schools],
                                "Total Students": [total_students],
                                "Total Budget": [total_budget],
                                "Average Math Score": [avg_math_score],
                                "Average Reading Score": [avg_read_score],
                                "% Passing Math": [percent_pass_math],
                                "% Passing Reading": [percent_pass_read],
                                "% Overall Passing Rate": [overall_percent_pass]})

District_Summary
District_Summary["Total Budget"]= District_Summary["Total Budget"].map("${:,.2f}".format)
District_Summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#complete_data = school_summary(complete_data)
group_schools = complete_data.groupby("school_name")
school_summary = pd.DataFrame({"School Name":school_data["school_name"], "School Type":school_data["type"],

                            "Total Budget":school_data["budget"]})
#rename table and values
school_summary = school_summary.set_index("School Name")
school_summary = school_summary.sort_values("School Name")
#school type 
school_summary["School Type"] = group_schools["type"]
#total students                               
school_summary["Total Students"] = group_schools["Student ID"].count()
#per student budget                                
per_student_budget=school_summary["Budget Per Student"] = school_summary["Total Budget"] / school_summary["Total Students"]
#average math score 
avg_math=school_summary["Average Math Score"] = group_schools["math_score"].mean()
#average reading score                                
avg_read=school_summary["Average Reading Score"] = group_schools["reading_score"].mean()
#% passing math 
math_pass = complete_data[complete_data['math_score'] >= 70].groupby('school_name')['Student ID'].count()/total_students                          
#% passing reading 
read_pass = complete_data[complete_data['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/total_students 
# overall passing rate
overall_rate = complete_data[(complete_data['reading_score'] >= 70) & (complete_data['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/total_students

#school summary dataframe 
school_summary = pd.DataFrame({"Total Schools": total_schools,
                                "Total Students": total_students,
                                "Total Budget": total_budget,
                                "Budget Per Student" : per_student_budget,
                                "Average Math Score": avg_math,
                                "Average Reading Score": avg_read,
                                "% Passing Math": math_pass,
                                "% Passing Reading": read_pass,
                                "% Overall Passing Rate": overall_rate})
school_summary = pd.DataFrame(school_summary)
#formatting 
school_summary["Total Budget"]= school_summary["Total Budget"].map("${:,.2f}".format)
school_summary['Budget Per Student'] = school_summary['Budget Per Student'].map('${:,.2f}'.format)
school_summary['% Passing Math'] = school_summary['% Passing Math'].map('{:,.2%}'.format)
school_summary['% Passing Reading'] = school_summary['% Passing Reading'].map('{:,.2%}'.format)
school_summary['% Overall Passing Rate'] = school_summary['% Overall Passing Rate'].map('{:,.2%}'.format)


school_summary

```
