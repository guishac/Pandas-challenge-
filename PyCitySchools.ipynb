```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
complete_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
school_total = len(complete_data['school_name'].unique())
# Calculate the total number of students
student_total = complete_data['student_name'].count()
# Calculate the total budget
budget_total = sum(complete_data['budget'].unique())
# Calculate the average math score
math_avg = complete_data['math_score'].mean()
# Calculate the average reading score
reading_avg = complete_data['reading_score'].mean()
# Calculate the overall passing rate
overall_pass = (math_avg + reading_avg)/2
# Calculate the percentage of students with a passing math score (70 or greater)
math_pass = (complete_data[complete_data['math_score']>=70]['student_name'].count()/student_total)*100
# Calculate the percentage of students with a passing reading score (70 or greater)
reading_pass = (complete_data[complete_data['reading_score']>=70]['student_name'].count()/student_total)*100

# Create a dataframe to hold the above results and formatting
district = {
    'Total Schools':school_total,
    'Total Student':'{:,}'.format(student_total),
    'Total Budget':'${:,.2f}'.format(budget_total),
    'Average Math Score':math_avg,
    'Average Reading Score':reading_avg,
    '% Passing Math':math_pass,
    '% Passing Reading':reading_pass,
    '% Overall Passing Score':[overall_pass],  
}

district_summary = pd.DataFrame(district)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Grouped our complete data frame by school name 
group_school = complete_data.groupby(['school_name'])
# Calculate the total student for each school
total_student = group_school.size()
# Get the school type for each school
school_type = group_school['type'].first()
# Calculate the total budget for each school
total_budget = group_school['budget'].first()
# Calculate the budget per student for each school
per_student_budget = budget_total/student_total
# Calculate the average math score for each school
math_avg = group_school['math_score'].mean()
# Calculate the average reading score for each school
reading_avg = group_school['reading_score'].mean()
# Calculate the percentange of passing math score for each school
group_math_pass = complete_data[complete_data['math_score']>=70].groupby(['school_name']).size()
percent_pass_math = (group_math_pass/student_total)*100
# Calculate the percentange of passing math score for each school
group_pass_reading = complete_data[complete_data['reading_score']>=70].groupby(['school_name']).size()
percent_pass_reading = (group_pass_reading/student_total)*100
# Calculate the overall passing score for each school
percent_overall_pass = (percent_pass_math + percent_pass_reading)/2

# Create a dataframe to hold the above results
school={
    'School Type': school_type,
    'Total Students':student_total,
    'Total School Budget': budget_total,
    'Per Student Budget': per_student_budget,
    'Average Math Score': math_avg,
    'Average Reading Score': reading_avg,
    '% Passing Math': percent_pass_math,
    '% Passing Reading': percent_pass_reading,
    '% Overall Passing Rate': percent_overall_pass,
}
school_summary = pd.DataFrame(school)
# Create a copy of school summary data frame before formatting to be able to use the numeric data on original data frame later
display_school_summary = school_summary.copy()
# Formatting the display data frame
display_school_summary['Per Student Budget'] = display_school_summary['Per Student Budget'].map('${:,.2f}'.format)
display_school_summary['Total School Budget'] = display_school_summary['Total School Budget'].map('${:,.2f}'.format)
display_school_summary.index.name = None
school_summary.head()
```

```python
top_performing = display_school_summary.sort_values(by='% Overall Passing Rate',ascending=False)
top_performing.head()
```
